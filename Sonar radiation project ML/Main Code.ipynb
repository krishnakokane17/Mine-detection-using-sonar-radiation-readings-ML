```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Data collection and data processing.
```

```python
#loading the dataset to a pandas data frame.
sonar_data = pd.read_csv("Copy of sonar radiation data.csv", header= None)
```

```python
sonar_data.head()
```

```python
#Number of rows and columns
sonar_data.shape
```

```python
#Describe gives the statistical measures of the data.
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

# M --> Mine R --> Rock

```python
sonar_data.groupby(60).mean()
```

```python
#Separating Data and Labels.
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

```python
model = LogisticRegression()
```

```python
#Training Logistic Regression Moddel with training data.
model.fit(X_train, Y_train)
```

```python
#Model Evaluation.
#Accuracy on training data.
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data: ", training_data_accuracy)
```

```python
#Accuracy on test data.
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on test data: ",test_data_accuracy)
```
